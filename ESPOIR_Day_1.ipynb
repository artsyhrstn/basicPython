# Exploratory Data Analysis


Exploratory Data Analysis refers to the process of conducting initial investigations on data, often with the goal of better understanding the characteristics of a specific dataset. EDA is performed using summary statistics and simple graphical techniques to examine the structure of the data at hand.

EDA typically involves:

1. Checking data by examining the structure/form of the data.
2. Checking data by looking at statistical summaries.
3. Checking data by visualizing the data.

Some simple tools in pandas that can be used for EDA are as follows:

### describe()

The `describe()` method displays 8 summary descriptive statistics. By default, it shows a summary for numeric columns.

The summary statistics include:

- Count: the number of rows in the dataframe.

- 50th Percentile (Q2/Median)
- 75th Percentile (Q3)
- Maximum Value: the largest value in the entire dataset.

These statistics provide an initial overview of the central tendency, spread, and distribution of the data.

These tools are useful for getting a quick sense of the data's structure and key statistical measures, allowing for a more informed approach to subsequent analysis.

## Conditional Subsetting

In addition to using .loc and .iloc, we can perform subsetting based on specific conditions. For example, in the "rice" dataframe, if we want to retrieve some data based on the following conditions:

1. Transactions that occurred at the supermarket: .format == 'supermarket'
2. Transactions with products priced greater than or equal to 200,000: .unit_price >= 200000
3. Transactions where the quantity is not equal to 0: .quantity != 0

The syntax for conditional subsetting is as follows:

```python
df[df['column_name'] <comparison_operator> <value>]
```

or

```python
df[df.column_name <comparison_operator> <value>]
```

Where the `<comparison_operator>` can be, for example, ==, !=, >, >=, <, <=.

````python
# format supermarket
rice[rice['format']=='supermarket']
````

````python
# unit_price >= 200000
rice[rice['unit_price']>=200000]
````

## Multiple Conditions

We can also use the & (AND) and | (OR) operators to perform subsetting with more than one condition. For example, if we want to view sales data from an employee named Moana with an amount greater than 5000, we can use the syntax:

```python
sales[(sales.salesperson == 'Moana') & (sales.amount > 5000)]
```

For subsetting with more than one condition, each condition is placed within parentheses (), or it can be written with the following syntax:

```python
df[(condition1) operator (condition2) operator (condition3), and so on...]
```

Points to note:

- AND operator: All conditions must be satisfied in one row for it to appear.
- OR operator: If any one condition is satisfied, the row will appear.

Example:

```python
cond1 = condition1
cond2 = condition2

df[(cond1) & (cond2)]  # Using AND operator
df[(cond1) | (cond2)]  # Using OR operator
```

❓ Lakukan subsetting untuk mengambil semua transaksi di supermarket yang unit_pricenya di atas 200,000.?

````python
# code here
rice[(rice['format'] == 'supermarket') & (rice['unit_price']>200000)]
````

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature (Independen) : {X.shape}') # Mengetahui dimensi data
print(f'Kelas:{set(y)} ') # Kelas pada target label
```

```python
from sklearn.model_selection import (train_test_split)

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3, # Testing di set menjadi 30% dari sekumpulan data yang dimiliki
                                                    random_state=0)

# Menghasilkan 4 kumpulan data, 4 buah variabel
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4) # Decision 4 tree yang dihasilkan memiliki kedalaman maksimum 4 layer

model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85 # Bisa disesuaikan sesuai kebutuhan
plt.subplots(figsize=(10,10))
tree.plot_tree(model, fontsize=10)
plt.show
```

```python
# Evaluasi Model
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
